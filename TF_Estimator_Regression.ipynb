```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv('cal_housing_clean.csv')
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop('medianHouseValue',axis=1), data['medianHouseValue'],
                                                    test_size=0.3, random_state=42)
```

```python
data.describe().T
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc=MinMaxScaler()
```

```python
sc.fit(X_train)
```

```python
X_train_tr=sc.transform(X_train)
X_test_tr=sc.transform(X_test)
```

```python
X_train=pd.DataFrame(data=X_train_tr,columns=data.columns[:-1],index=X_train.index)
X_test=pd.DataFrame(data=X_test_tr,columns=data.columns[:-1],index=X_test.index)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
name_of_cols='medAge totRooms totBeds pop households medIncome'.split()
```

```python
import tensorflow as tf
```

```python
for el,col in zip(name_of_cols,data.columns[:-1]):
    exec("{}=tf.feature_column.numeric_column('{}')".format(el,col))
```

```python
feat_cols=[medAge, totRooms, totBeds, pop, households, medIncome]
```

```python
batch_size=10
num_of_epochs=500
```

```python
train_fn=tf.estimator.inputs.pandas_input_fn(X_train,
                                             y_train,
                                             batch_size=batch_size,num_epochs=num_of_epochs,shuffle=True)
```

```python
reg_model=tf.estimator.DNNRegressor([10,10,10],feat_cols)
reg_model.train(input_fn=train_fn,steps=20000)
```

```python
test_fn=tf.estimator.inputs.pandas_input_fn(X_test,batch_size=10,num_epochs=1,shuffle=False)
```

```python
predictions=reg_model.predict(input_fn=test_fn)
```

```python
predictions
```

```python
preds=list(predictions)
```

```python
lop=[]
for pred in preds:
    lop.append(pred['predictions'])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,lop)**.5
```
