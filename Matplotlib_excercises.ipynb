```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid.inset_locator import inset_axes
```

```python
x=np.linspace(0,5,11)
```

```python
color='r g b k y springgreen salmon gold fuchsia'.split()
ls=['-', '--', '-.', ':','-', '--', '-.', ':','steps']
fig,axes=plt.subplots(nrows=3,ncols=3,figsize=(12,8))
plt.tight_layout()
for i,row in enumerate(axes):
    for j,col in enumerate(row):
        a=axes.shape[0]*i+j
        col.plot(x,x**a,color=color[a],ls=ls[a],markersize=3,
                 marker='o',markerfacecolor='green',label='X^{}'.format(a))
        col.set_xlabel('X axis')
        col.set_ylabel('Y axis')
        col.set_title('X^{}'.format(a))
        col.legend(loc=10)
        inner_ax=inset_axes(col,'30%','30%',loc=2)
        inner_ax.plot(x,x**a,color=color[a],ls=ls[a])
```
